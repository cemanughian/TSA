the box test - you have to correct the DF for every param u estimate. these resids are produced by a model in which u estimate 2 params, so we have to subtract 2 df from the distribution of the test statistic.



R knows what the fit is and that is has 2 params but it didnt adjust for the number of parameters that it itself estimated in the model, so we have to feed the number of parameters that we estimated manually. 

```r
Box.test(resids, lag=10, fitdf=2)
```
