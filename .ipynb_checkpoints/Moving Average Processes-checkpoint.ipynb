We can plot the time series against itself but lagged one point back 


```r
plot(x=zlag(ma1.2.s),y=ma1.2.s)
```

This is visually inspecting how correlated observation are with the one day lag. loosely positively correlatation between lagged and timesries itself, this shows cor was positive and that theta was negative. *** review  

How many points on the graph? n-1

What would happen if we lag two days back? 

```r
plot(x=zlag(ma1.2.s,2),y=ma1.2.s)
```

We see no correlation for all possible pairs of time points two days apart. the dependence between the obs and the one day old obs is evident in the previous graph. going two days back the dependence vanishes as the theory suggests it should.
