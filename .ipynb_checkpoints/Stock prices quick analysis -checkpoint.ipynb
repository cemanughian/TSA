Steady increase, variations, and minor and major peaks. 

Is this an example of a stationary time series? I dont think so 

The mean and variance needs to be constant. The mean is definitely changing. 
Stochastic process driven by correlation, not because the mean is increasing in a predicble way. There are hypothesis tests for stationarity. Just by looking, possibly on the variance side we are fine, ofc there are some bursts of variation. On the mean side of the process it looks like theres a change of mean of the process over time, and this is inconsistent with a stationary process behavior. 

```r
plot(AMZN$AMZN.Close)
```

Is this a stationary process?

If you focus from august to now, clearly its not stationary. Very steep incline and then stabilization is very hard to believe its a stationary process. If things somehow came down quickly after the climb you could say there was a stochastic trend and the process changed direction but the correlation is driving it back down, but thats not the case.
On the variance side, it wobbles up and down, it seems like theres a change in variability over time, it seems inconsistent with stationary on the variance side but definitely on the mean side.
