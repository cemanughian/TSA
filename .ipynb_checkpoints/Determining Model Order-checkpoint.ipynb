```r
data(ar1.s)
```

```r
plot(ar1.s)
```

This the AR(1) look different from MA?

remember the lingering influence changes the shape of the process. 

lets check acf and maybe we can define it as ma

```r
acf(ar1.s)
```

You can see how we have a classical pattern of ACF values 

Why is this classical?

Would u say theres any hope that this is MA process based on this graph? Do you see anything dying off at zero and never popping up in a systematic fashion again? 


we have these systematic patterns of nonzero blocks which is what we usually observe when we look at acf of AR or mixed processes. we know immediately its not MA. Our next hope is to be pure AR. so lets calculate PACF values. 

```r
pacf(ar1.s)
```

These should behave the same way for AR as the ACF for MA. 

we see its pretty clear its pure AR(1) as evidenced by the huge PACF value at lag 1 and the rest being within the blue band.

<b> extended acf consultation

```r
eacf(arma11.s)
```

doing a gridsearch MA up to 14 and AR up to 7 to estimate a hypothestical number of parameters


here X means the model is not a model not supported by data, o's are models that are supported by the data. remember we look for most parsimonious - closest to upper LHS corner (1,1 is the one with an o that is closest to upper LHS, which is actually the true model. 

so (1,1) is the most parsimonious model with a good fit

you might be curious, what would autoarima say which is looking at the bets model selection in a completely different way. its not looking at vanishing pacf values or acf, or estimating subtracting and looking for vanishing parts. it has an analytical decision making tools. chooses model with Min AIC.

```r
library(forecast)
```

```r
auto.arima(arma11.s)
```
